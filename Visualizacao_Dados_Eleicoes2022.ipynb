```python
import basedosdados as bd
import pandas as pd
```

```python
df = bd.read_table(dataset_id='br_poder360_pesquisas',
table_id='microdados',
billing_project_id="eleicoes-2022-hg")
```

```python
df.head()
```

```python
dados2022 = df[(df.instituto == 'Datafolha') 
      & (df.ano == 2022) 
      & (df.cargo == 'presidente') 
      & (df.sigla_uf.isnull())
      & (df.tipo == 'estimulada')
      & (df.turno == 1)
      & (df.descricao_cenario.str.contains('cenário 1')==True)
      & ((df.nome_candidato == 'Bolsonaro') | (df.nome_candidato == 'Ciro') | (df.nome_candidato == 'Lula') | (df.nome_candidato == 'Simone Tebet'))
      ].sort_values(by='data')
```

```python
dados2022.head()
```

```python
pd.options.plotting.backend = "plotly"
```

```python
dados2022.plot(x='data', y='percentual', color='nome_candidato' )
```
